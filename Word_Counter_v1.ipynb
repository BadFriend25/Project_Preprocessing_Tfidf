```python
import PyPDF2

def count_words_in_pdf(pdf_path):
    with open(pdf_path, 'rb') as f:
        pdf_reader = PyPDF2.PdfReader(f)
        text = ""
        for page in range(len(pdf_reader.pages)):
            page_obj = pdf_reader.pages[page]
            text += page_obj.extract_text()

    words = text.split()
    word_count = len(words)
    return word_count

pdf_path = "D:\Kementerian Kominfo_ Situs Elaelo Pengganti X_Twitter Bukan Buatan Pemerintah.pdf"
word_count = count_words_in_pdf(pdf_path)
print(f"The number of words in the PDF file is: {word_count}")
```

```python
import PyPDF2
import string

def get_all_words_in_pdf(pdf_path):
    with open(pdf_path, 'rb') as f:
        pdf_reader = PyPDF2.PdfReader(f)
        text = ""
        for page in pdf_reader.pages:
            text += page.extract_text()

    # Remove punctuation
    text_no_punct = text.translate(str.maketrans('', '', string.punctuation))
    words = text_no_punct.split()
    unique_words = list(set(words))
    unique_words.sort()
    return unique_words

pdf_path = "D:\Kementerian Kominfo_ Situs Elaelo Pengganti X_Twitter Bukan Buatan Pemerintah.pdf"
all_words = get_all_words_in_pdf(pdf_path)
print("All unique words in the PDF file (including stop words), sorted alphabetically:")
for word in all_words:
    print(word)
```

```python
import PyPDF2
import string
from collections import Counter

def get_all_words_in_pdf(pdf_path):
    with open(pdf_path, 'rb') as f:
        pdf_reader = PyPDF2.PdfReader(f)
        text = ""
        for page in pdf_reader.pages:
            text += page.extract_text()

    # Remove punctuation
    text_no_punct = text.translate(str.maketrans('', '', string.punctuation))
    words = text_no_punct.split()
    word_freq = Counter(words)
    return word_freq

pdf_path = "D:\Samkok - Luo Guanzhong.pdf"
word_freq = get_all_words_in_pdf(pdf_path)
print("Word frequencies:")
for word, freq in word_freq.most_common():
    print(f"{word}: {freq}")
```
